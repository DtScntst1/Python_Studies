```python
import csv
import random

# CSV dosyasının adını oluşturmak ve çağırmak için değişkeni ata.
dosya_adi = "School.csv"
basliklar = ["","Country", "Region", "City", "School", "Class","Max #of Selection","Min #of Student","Max #of Student"]
sayac = 0
# Ülkeler ve bölgeler veritabanı
ulke_bolgeleri = {
    "Türkiye": ["Ege Bölgesi", "Karadeniz Bölgesi", "Doğu Anadolu Bölgesi"],
    "Almanya": ["Saksonya", "Brandenburg", "Baden Württemberg"],
    "Amerika": ["Alabama", "Alaska", "Arizona"]
}

bolge_sehirleri = {
    "Ege Bölgesi": ["İzmir", "Manisa", "Aydın"],
    "Karadeniz Bölgesi": ["Trabzon", "Ordu", "Giresun"],
    "Doğu Anadolu Bölgesi": ["Malatya", "Elazığ", "Erzurum"],
    "Saksonya": ["Leipzig", "Chemnitz", "Zwickau"],
    "Brandenburg": ["Cottbus", "Brandenburg an der Havel", "Frankfurt"],
    "Baden Württemberg": ["Konstanz", "Freiburg", "Karlsruhe"],
    "Alabama": ["Abbeville", "Adamsville", "Addison"],
    "Alaska": ["Anchorage", "Fairbanks", "Juneau"],
    "Arizona": ["Bullhead City", "Chandler", "Flagstaff"]
}

okul_isimleri = {
    "Türkiye":["Atatürk", "Cumhuriyet", "Mustafa Kemal","İnönü", "Gazi", "Mehmet Akif Ersoy", "İstiklal", "Fatih", "Yunus Emre"],
    "Almanya":["Goethe", "Martin Luther", "Heinrich Heine", "Schiller", "Waldorf", "Schule am See", "Humboldt", "Lessing", "Einstein"],
    "Amerika":["Lincoln", "Roosevelt", "John F. Kennedy" ,"Central","Parkview", "Martin Luther King Jr.", "Washington", "Jefferson", "Maplewood"]
}


# CSV dosyasını açın ve yazma modunda bir CSV yazıcı oluşturun
with open(dosya_adi, "w", newline="", encoding="utf-8") as dosya:
    csv_yazici = csv.writer(dosya, delimiter=";")

    # CSV dosyasının başlık satırını yazın
    csv_yazici.writerow(basliklar)

    # Her bölge için rastgele okul verileri oluşturun ve CSV dosyasına yazın
    for ulke, bolgeler in ulke_bolgeleri.items():
        for bolge in bolgeler:
            for ulke_, okullar in okul_isimleri.items():
                secilen_okullar = random.sample(okul_isimleri[ulke], 3)
                for okul in secilen_okullar:
                    if ulke == ulke_:
                        sayac += 1
                        sehir = random.choice(bolge_sehirleri[bolge])
                        sinif = 12
                        max_selction = 3
                        min_student = 15
                        max_student = 25

                        # Oluşturulan verileri CSV dosyasına yazın
                        csv_yazici.writerow([sayac, ulke, bolge, sehir, okul, sinif, max_selction, min_student, max_student])

print(f"'{dosya_adi}' adlı CSV dosyası oluşturuldu.")
```

```python
import csv
import random
from faker import Faker

fake = Faker()

okunan_dosya_adi_1 = "SchoolRegistration.csv"
okunan_dosya_adi_2 = "School.csv"
yazilacak_dosya_adi = "Student.csv"
sayac = 0


# Türkçe erkek isimleri
turkce_erkek_isimleri = ["Ahmet", "Mehmet", "Mustafa", "Ali", "Hüseyin", "Emre", "Kadir", "Sedat", "Yusuf"]

# Türkçe kadın isimleri
turkce_kadin_isimleri = ["Ayşe", "Fatma", "Zeynep", "Sibel", "Nur", "Elif", "Gizem", "Seda", "Ebru", "Dilara"]

# Almanca erkek isimleri
almanca_erkek_isimleri = ["Lukas", "Maximilian", "Benjamin", "Felix", "Leon", "Paul", "Jan", "Niklas", "Tobias", "Jonas"]

# Almanca kadın isimleri
almanca_kadin_isimleri = ["Sophie", "Lena", "Mia", "Hannah", "Emma", "Marie", "Laura", "Lea", "Emily", "Sarah"]

# Amerikan erkek isimleri
amerikan_erkek_isimleri = ["James", "John", "Robert", "Michael", "William", "David", "Joseph", "Daniel", "Charles", "Matthew"]

# Amerikan kadın isimleri
amerikan_kadin_isimleri = ["Mary", "Jennifer", "Linda", "Patricia", "Elizabeth", "Susan", "Jessica", "Sarah", "Karen"]

# Soyadları
turkce_soyadlari = ["Yılmaz", "Kaya", "Öztürk", "Demir", "Yıldırım", "Çelik", "Koç", "Aydın", "Arslan", "Türk"]

almanca_sayadlari = ["Schmidt", "Müller", "Kaiser", "Schulz", "Wagner", "Becker", "Hoffmann", "Schneider", "Fischer"]

amerikan_soyadlari = ["Smith", "Johnson", "Brown", "Jones", "Davis", "Miller", "Wilson", "Moore", "Taylor", "Anderson"]


# Okul isimlerini "School.csv" dosyasından çekme

ogrenci_numarasi = []
with open(okunan_dosya_adi_1, "r", newline="", encoding="utf-8") as okunan_dosya:
    csv_okuyucu = csv.reader(okunan_dosya, delimiter=";")
    basliklar = next(csv_okuyucu)  # Başlık satırını oku

    for sutun in csv_okuyucu:
        student = sutun[4]
        ogrenci_numarasi.append(student)


ulkeler = []
with open(okunan_dosya_adi_2, "r", newline="", encoding="utf-8") as okunan_dosya:
    csv_okuyucu = csv.reader(okunan_dosya, delimiter=";")
    basliklar = next(csv_okuyucu)  # Başlık satırını oku

    for sutun in csv_okuyucu:
        ulke = sutun[1]
        ulkeler.append(ulke)

# "Student.csv" dosyasına öğrenci verilerini yazma
with open(yazilacak_dosya_adi, "w", newline="", encoding="utf-8") as yazilacak_dosya:
    csv_yazici = csv.writer(yazilacak_dosya, delimiter=";")

    basliklar = ["", "Student", "Gender", "Name", "Last_name"]

    csv_yazici.writerow(basliklar)

    # Her okul için rastgele öğrenci verileri oluşturup yazma
    for student in range(len(ogrenci_numarasi)):  # Her okul için 30 öğrenci oluştur
            for ulke in ulkeler:
                cinsiyet = fake.random_element(["Male", "Female"])
                if ulke == "Türkiye":
                    if cinsiyet == "Male":
                        name = random.choice(turkce_erkek_isimleri)
                        surname = random.choice(turkce_soyadlari)
                    else:
                        name = random.choice(turkce_kadin_isimleri)
                        surname = random.choice(turkce_soyadlari)
                elif ulke == "Almanya":
                    if cinsiyet == "Male":
                        name = random.choice(almanca_erkek_isimleri)
                        surname = random.choice(almanca_sayadlari)
                    else:
                        name = random.choice(almanca_kadin_isimleri)
                        surname = random.choice(almanca_sayadlari)
                else:
                    if cinsiyet == "Male":
                        name = random.choice(amerikan_erkek_isimleri)
                        surname = random.choice(amerikan_soyadlari)
                    else:
                        name = random.choice(amerikan_kadin_isimleri)
                        surname = random.choice(amerikan_soyadlari)

                    sayac += 1

                        # Oluşturulan verileri CSV dosyasına yaz
                    csv_yazici.writerow([sayac, student + 1, cinsiyet, name, surname])

print(f"'{yazilacak_dosya_adi}' adlı yeni CSV dosyası oluşturuldu.")
```
